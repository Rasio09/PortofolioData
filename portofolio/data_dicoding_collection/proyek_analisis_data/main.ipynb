<h3> RFM Analysis 

RFM analysis merupakan salah satu metode yang umum digunakan untuk melakukan segmentasi pelanggan (mengelompokkan pelanggan ke dalam beberapa kategori) berdasarkan tiga parameter, yaitu recency, frequency, dan monetary.

1. Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.

2. Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.

3. Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
# mengelompokkan data berdasarkan customer_id, memastikan bahwa setiap baris hasilnya adalah representasi dari satu pelanggan.
rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_date": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "total_price": "sum" # menghitung jumlah revenue yang dihasilkan
})
# penamaan kolom tabel
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date #Mengubah kolom max_order_timestamp menjadi hanya tanggal tanpa waktu.
recent_date = orders_df["order_date"].dt.date.max() #engambil tanggal transaksi terakhir dalam orders_df untuk menghitung seberapa lama waktu telah berlalu sejak transaksi terakhir setiap pelanggan.
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days) #Menghitung recency, yaitu jumlah hari yang telah berlalu sejak transaksi terakhir setiap pelanggan.
 
rfm_df.drop("max_order_timestamp", axis=1, inplace=True) #enghapus kolom max_order_timestamp karena informasi tersebut sudah tidak diperlukan lagi setelah menghitung recency.
rfm_df.head() #cetak
```

```python
# mengidentifikasi best customer berdasarkan 3 parameter

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"] #cetak warna
 
'''sns.barplot() digunakan untuk membuat bar chart.
y="recency" dan x="customer_id" menunjukkan bahwa sumbu y adalah recency dan sumbu x adalah customer_id.
data=rfm_df.sort_values(by="recency", ascending=True).head(5) mengurutkan DataFrame berdasarkan nilai recency secara menaik (terkecil) dan memilih 5 pelanggan dengan recency terkecil.
ax[0] menentukan bahwa plot ini akan ditampilkan di subplot pertama.
set_title() menetapkan judul "By Recency (days)" untuk subplot pertama.
tick_params() digunakan untuk mengubah ukuran label pada sumbu x.'''

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

'''Sama seperti langkah sebelumnya, tetapi kali ini visualisasi dilakukan berdasarkan frekuensi (frequency), dengan memilih 5 pelanggan yang memiliki frekuensi transaksi tertinggi.
Data diurutkan berdasarkan frequency secara menurun (terbesar) dengan .sort_values(by="frequency", ascending=False).'''
 
sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

'''Sama seperti langkah sebelumnya, tetapi kali ini visualisasi dilakukan berdasarkan total pendapatan (monetary), dengan memilih 5 pelanggan yang menghasilkan pendapatan tertinggi.'''
 
sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 

plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)  #Menambahkan judul global di atas seluruh subplot dengan ukuran font 20
plt.show() #cetak
```
