<h3>EKSPLORASI DATA ORDER dan CUSTOMERS DATA

<h4> JOIN DATA

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id", #ini field penggabungannya bisa berbeda namanya
    right_on="customer_id" #ini field penggabungannya bisa berbeda namanya
)
orders_customers_df.head()
```

```python
# cek jumlah order berdasasrkan kota
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
# cek jumlah order berdasarkan state
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

```python
# cek jumlah order berdasarkan gender
orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

```python
# cek jumlah order berdasarkan kelompok usia
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
# mencetak data all untuk mengetahui keseluruhan
all_df = pd.merge(
    left=sales_product_df,
    right=orders_customers_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_df.head()
```

```python
# menjawab pertanyaan nomor 1
# rule M adalah frekuensi bulanan (Month)
# on adalah untuk mengemlompokkan data mana
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
# ini untuk menuliskan data dari urutan tahun - bulan
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
# reset_index() ini untuk mengubah index menjadi sebuah kolom biasa
monthly_orders_df = monthly_orders_df.reset_index()
# ini untuk merubah namanya
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
monthly_orders_df.head()
```

```python
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%B') #mengubah format order date menjadi nama bulan
 
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
 
plt.figure(figsize=(10, 5)) 
plt.plot(monthly_orders_df["order_date"], monthly_orders_df["order_count"], marker='o', linewidth=6, color="#72BCD4") 
plt.title("Number of Orders per Month (2021)", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```
